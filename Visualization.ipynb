```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
drive.mount('/content/drive')

data = pd.read_csv("/content/drive/MyDrive/碩士論文/論文程式/Electrical Grid.csv")
data_x = data.drop(['stab','stabf'],axis=1)
data_y = data['stab']
```

```python
data_x.describe()
```

```python
plt.figure(figsize=(11,11))
corr = data.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, annot=True,fmt='.2f',mask=mask, robust=True, center=0,square=True, linewidths=.6,cmap='YlOrRd_r') #cmap='YlGnBu'
plt.show()
```

```python
# pie chart
labels = ['unstable', 'stable']
size = data['stabf'].value_counts()
colors = ['crimson','seagreen']
explode = [0, 0.05]

print(data['stabf'].value_counts())
plt.style.use('seaborn-deep')
plt.rcParams['figure.figsize'] = (5 ,5)
plt.pie(size, labels =labels , colors = colors, explode = explode, autopct = "%.2f%%", shadow = False,textprops = {"fontsize" : 13})
plt.axis('off')
plt.legend()
plt.show()
```
