```python
import pandas as pd
import numpy as np

from sklearn.model_selection import KFold
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR

data = pd.read_csv("/content/drive/MyDrive/碩士論文/論文程式/Electrical Grid.csv")

data_x = data.drop(['stab','stabf'],axis=1)
data_y = data['stab']

scaler = MinMaxScaler()
scaled_x = scaler.fit_transform(data_x)
scaled_x = pd.DataFrame(scaled_x, columns = data_x.columns)

kf = KFold(n_splits=10, shuffle=True, random_state=42)
cnt = 1

RMSE_set_DT , RMSE_set_GBDT ,RMSE_set_KNN ,RMSE_set_MLP,RMSE_set_SVR,RMSE_set_LR =[],[],[],[],[],[]
for train_index, valid_index in kf.split(scaled_x, data_y): # split()  method generate indices to split data intSo training and test set.
    print('fold',cnt)
    train_x = scaled_x.iloc[train_index,:]
    train_y = data.iloc[train_index,:]['stab']
    valid_x = scaled_x.iloc[valid_index,:]
    valid_y = data.iloc[valid_index,:]['stab']

    DT_pred = DecisionTreeRegressor(max_depth=30,min_samples_leaf=10).fit(train_x,train_y).predict(valid_x)
    SVR_pred = SVR(C=1,epsilon=0.025).fit(train_x,train_y).predict(valid_x)
    LR_pred = LinearRegression().fit(train_x,train_y).predict(valid_x)
    GBDT_pred = GradientBoostingRegressor(n_estimators = 100).fit(train_x,train_y).predict(valid_x)
    KNN_pred = KNeighborsRegressor(n_neighbors = 10).fit(train_x,train_y).predict(valid_x)
    MLP_pred = MLPRegressor(hidden_layer_sizes=(100,100)).fit(train_x,train_y).predict(valid_x)

    RMSE_set_DT.append( np.sqrt( mean_squared_error(valid_y,DT_pred)))
    RMSE_set_SVR.append( np.sqrt( mean_squared_error(valid_y,SVR_pred)))
    RMSE_set_GBDT.append( np.sqrt( mean_squared_error(valid_y,GBDT_pred)))
    RMSE_set_KNN.append( np.sqrt( mean_squared_error(valid_y,KNN_pred)))
    RMSE_set_MLP.append( np.sqrt( mean_squared_error(valid_y,MLP_pred)))
    RMSE_set_LR.append( np.sqrt( mean_squared_error(valid_y,LR_pred)))
    cnt+=1

print("SVR",RMSE_set_SVR)
print("mean",np.mean(RMSE_set_SVR))

print("DT",RMSE_set_DT)
print("mean",np.mean(RMSE_set_DT))

print("GBDT",RMSE_set_GBDT)
print("mean",np.mean(RMSE_set_GBDT))

print("KNN",RMSE_set_KNN)
print("mean",np.mean(RMSE_set_KNN))

print("MLP",RMSE_set_MLP)
print("mean",np.mean(RMSE_set_MLP))

print('LR',RMSE_set_LR)
print("mean",np.mean(RMSE_set_LR))
```
